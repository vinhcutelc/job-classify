# 1. Tiền xử lý dữ liệu

```python
## xử lý dữ liệu lương
job1 = job[job.salary.notnull()]
for i in range(0,len(job1.salary)):
    job1.salary[i]= job1.salary[i].split('-')[0]
    if 'triệu' in job1.salary[i]:
        price = job1.salary[i].split(' ')[0]
        job1.salary[i] = round(int(price)*1000000,1)
    else:
        job1.salary[i] = round(int(price)*1000000,1)
job1.head()
```

```python
## tiền xử lý dữ liệu chuyên ngành
job2 = job1[job1.keyword.notnull()]
for i in range(0,len(job2.keyword)):
    if '/' in job2.keyword[i]:
        job2.keyword[i]= job2.keyword[i].split('/')[0]

for i in range(0,len(job2.keyword)):
    if 'IT' in job2.keyword[i]:
        job2.keyword[i]= 'CNTT'
    if 'Bảo trì' in job2.keyword[i]:
        job2.keyword[i] = 'CNTT'
    if 'Kinh doanh' in job2.keyword[i]:
        job2.keyword[i] = 'Kinh doanh'
    if 'Dịch vụ' in job2.keyword[i]:
        job2.keyword[i] = 'Dịch vụ'
    if 'Kế toán' in job2.keyword[i]:
        job2.keyword[i] = 'Hành chính'
    if 'Thiết kế' in job2.keyword[i]:
        job2.keyword[i] = 'Thiết kế'
    if 'Kiến trúc' in job2.keyword[i]:
        job2.keyword[i] = 'Kiến trúc'
    if 'Xây dựng' in job2.keyword[i]:
        job2.keyword[i] = 'Xây dựng'
    if 'Logistic' in job2.keyword[i]:
        job2.keyword[i] = 'Logistic'
    if 'Vận tải' in job2.keyword[i]:
        job2.keyword[i] = 'Logistic'
    if 'Nhân sự' in job2.keyword[i]:
        job2.keyword[i] = 'Hành chính'
    if 'Hành chính' in job2.keyword[i]:
        job2.keyword[i] = 'Hành chính'
    if 'Sản xuất' in job2.keyword[i]:
        job2.keyword[i] = 'Sản xuất'
    if 'Bán' in job2.keyword[i]:
        job2.keyword[i] = 'Kinh doanh'
    if 'Ngân hàng' in job2.keyword[i]:
        job2.keyword[i] = 'Ngân hàng'
    if 'Thời trang' in job2.keyword[i]:
        job2.keyword[i] = 'Thiết kế'
    if 'Spa' in job2.keyword[i]:
        job2.keyword[i] = 'Dịch vụ'
    if 'Du lịch' in job2.keyword[i]:
        job2.keyword[i] = 'Dịch vụ'
        
job2.head()
```

```python
## tiền xử lý dữ liệu yêu cầu kinh nghiệm
job3 = job2[job2.experience_required.notnull()]
for i in range(0,len(job3.experience_required)):
    if 'Không yêu cầu kinh nghiệm' in job3.experience_required[i]:
        job3.experience_required[i]= '0'
    if 'Dưới 1' in job3.experience_required[i]:
        job3.experience_required[i]= '0'
job3.head()
```

```python
## tiền xử lý dữ liệu hình thức làm việc
job4 = job3[job3.work_type.notnull()]
for i in range(0,len(job4.work_type)):
    if 'Toàn thời gian' in job4.work_type[i]:
        job4.work_type[i]= '0'
    if 'Remote' in job3.work_type[i]:
        job4.work_type[i]= '1'
    if 'Làm việc từ xa' in job3.work_type[i]:
        job4.work_type[i]= '1'
    if 'Bán thời gian' in job3.work_type[i]:
        job4.work_type[i]= '2'
    if 'Thực tập' in job3.work_type[i]:
        job4.work_type[i]= '2'
job4.head()
```

```python
## tiền xử lý dữ liệu yêu cầu giới tính
job5 = job4[job4.gender.notnull()]
for i in range(0,len(job5.gender)):
    if 'Không yêu cầu' in job5.gender[i]:
        job5.gender[i]= '0'
    if 'Nam' in job5.gender[i]:
        job5.gender[i]= '1'
    if 'Nữ' in job5.gender[i]:
        job5.gender[i]= '2'
job5.head()
```

```python
## xuất file dataset sau khi xử lý
job5.to_csv('final_set_job.csv',encoding="utf-8-sig",index=False)
```

## 3. Học máy

### 3.1. Linear regression

```python
## đọc file dataset
job = pd.read_csv('final_set_job.csv')
job.info()
job.dtypes
```

```python
job.describe()
```

```python
## định dạng dữ liệu
data_dummies = pd.get_dummies(job, drop_first=True)
data_dummies = data_dummies.astype(float)
dataset = data_dummies.values
dataset
```

```python
## tách dataset
X = dataset[:,1:5]
Y = dataset[:,0]
```

## Linear Regression

```python
X
```

```python
Y
```

```python
one = np.ones((X.shape[0], 1))
Xbar = np.concatenate((one, X), axis = 1)
```

```python
A = np.dot(Xbar.T, Xbar)
b = np.dot(Xbar.T,Y)
```

```python
A
```

```python
b
```

```python
A_dagger = np.linalg.pinv(A)
A_dagger
```

```python
w = np.dot(A_dagger, b)
w
```

```python
## dự doán thủ công
w_0 = w[0]
w_1 = w[1]
w_2 = w[2]
w_3 = w[3]
w_4 = w[4]
y1 = w_1/2 + w_2*0 + w_3*3 + w_4*1 + w_0
print(u'Với hình thức làm việc toàn thời gian, không yêu cầu giới tính,')
print( u'Dự đoán : %.2f (vnd), Mức lương thật: 40000000 (vnd)'  %(y1) )
print( u'Sai lệch : %.2f (vnd)'  %(40000000-y1) )
```

```python
## chia tỉ lệ dữ liệu
from sklearn import preprocessing
min_max_scaler = preprocessing.StandardScaler()
X_scale = min_max_scaler.fit_transform(Xbar)
X_scale
```

```python
## tách dataset để tiến hành train
from sklearn.model_selection import train_test_split
X_train, X_val_and_test, Y_train, Y_val_and_test = train_test_split(X_scale, Y, test_size=0.3)
X_val, X_test, Y_val, Y_test = train_test_split(X_val_and_test, Y_val_and_test, test_size=0.5)

print(X_train.shape, X_val.shape, X_test.shape, Y_train.shape, Y_val.shape, Y_test.shape)
```

## sklearn Linear Regression

```python
from sklearn.linear_model import LinearRegression
```

```python
reg = LinearRegression().fit(Xbar, Y)
```

```python
w_sklearn = reg.coef_
w_sklearn
```

```python
w
```

```python
Y_test
```

```python
Y_pre = reg.predict(X_test)
```

```python
pre = (1/2)*((Y_pre - Y_test)**2)
perc = np.abs((Y_pre - Y_test)/Y_test)
perc = perc*100
pre
```

```python
data = {'Mức lương dự đoán':Y_pre,
        'Mức lương thật':Y_test,
       '% sai lech': perc}
linear_pred = pd.DataFrame(data)
linear_pred
```

```python
linear_pred = linear_pred.sort_values(by=['% sai lech'])
linear_pred
```
